### Preprocessing

```python
import pandas as pd
import re
import matplotlib.pyplot as plt

df = pd.read_csv('movie_dataset_classification.csv',index_col=0)
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df["Release Year"] = df["Release Year"].astype(int)
```

```python
df = df[(df["Director"] != "Unknown") & (df["Cast"] != "Unknown") & (df["Cast"] != "[]")]
df.shape
```

```python
df.tail(50)
```

```python
def cast_cleaning(cast):
    cast = re.sub(r'[\'\"\[\]]', '', cast)
    cast = re.sub(r' \(.+\)', '', cast)
    cast = [actor for i, actor in enumerate(cast.split(', ')) if i < 3]
    cast_txt = ', '.join(cast)
    return cast_txt
```

```python
df["Cast"] = df["Cast"].apply(cast_cleaning)
```

### Doc embeddings

```python
# 3 = cast > split
# 4 = plot > tok

def tok_cast_plot(row):
    row[3] = row[3].split(",")
    row[4] = re.split('[\W\d_]+', row[4].lower())
    return row
```

```python
from gensim.models.doc2vec import TaggedDocument

data = list()
for row in df.iterrows():
    #row = list(row[1])
    row = tok_cast_plot(row[1].values)
    data.append(TaggedDocument(row[4],[row[1]]))
    
data
```

```python
import logging
import multiprocessing
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
cores = multiprocessing.cpu_count()
cores
```

```python
from gensim.models import Doc2Vec
model = Doc2Vec(data, vector_size=300, window=5, min_count=5, workers=cores)
```

```python
model.dv.most_similar("Conquest of Space")
```

```python
model.dv[0]
```

```python
X = [model.dv[i] for i in range(len(model.dv))]
```

```python
X[0]
```

```python
tags = model.dv.index_to_key
```

### Representation

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
vecs = pca.fit_transform(X)
```

```python
df_docs = pd.DataFrame(vecs)
df_docs['Title'] = tags
df_docs = df_docs.merge(df[['Title','Genre']],on='Title').drop_duplicates()
```
