```python
import pandas as pd

df = pd.read_csv("movie_plots_yts.csv", encoding='unicode_escape')
```

```python
df.head(60)
```

```python
df = df[~df['Plot'].isna()].drop_duplicates()
df
```

```python
import re

def genre_preprocessing(genre):
    genre = re.sub("[\[\]\']", "", genre)
    return genre
```

```python
df['Genres'] = df['Genres'].apply(genre_preprocessing)
```

```python
genres = set([genre.strip() for row in df['Genres'] for genre in row.split(",")])
genres
```

```python
genres_to_drop = ['Game-Show','Talk-Show','Mystery','News','Reality-TV']
mask = df.Genres.apply(lambda x: any(g in x for g in genres_to_drop))
df = df[~mask]
df.shape
```

```python
set([genre.strip() for row in df['Genres'] for genre in row.split(",")])
```

```python
df.to_csv("movie_plots_yts.csv", encoding='utf-8')
```
