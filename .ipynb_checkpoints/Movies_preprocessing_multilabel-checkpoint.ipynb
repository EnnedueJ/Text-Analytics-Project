```python
import pandas as pd
import re
```

```python
df = pd.read_csv('wiki_movie_plots_deduped.csv')
```

```python
df = df[df['Genre'] != 'unknown']
df.dropna(subset=['Plot'], inplace=True)
counts = df.Genre.value_counts()
counts = list(counts[counts < 30].index)
df = df[~df['Genre'].isin(counts)]
```

```python
conversion_dict = {
    "action": ["disaster", "martial arts", "spy", "superhero", "wuxia","action","masala","espionage","arts"],
    "adventure": ["pirate", "swashbuckler", "samurai"],
    "animation": ["cgi", "cutout", "live-action animated film", "stop motion", "animated", "computer-animated", "anime"],
    "comedy": ["buddy", "mockumentary", "parody", "slapstick"],
    "drama": ["docudrama", "melodrama", "biodrama", "bio-drama"],
    "historical": ["history", "historic", "alternate history", "period", "period piece", "biopic", "bio-pic", "biographical"],
    "horror": ["ghost", "monster", "vampire", "werewolf", "slash", "splatter", "zombie", "j-horror","supernatural"],
    "science fiction": ["dystopian", "dystopia", "post-apocalyptic", "steampunk", "tech noir", "utopian", "science-fiction", "scifi", "sci-fi", "space", "tokusatsu","fiction"],
    "thriller": ["mystery", "detective", "crime","suspense"],
    "musical": ["operetta"],
    "romance": ["love","romantic"],
    "western": ["cowboy"],
    "documentary": ["pseudo-documentary"],
    "fantasy":[],
    "sport":["sports","races","dance","biker"],
    "war":['ii','i'],
    "erotic":['ero','adult','erotic','sexploitation'],
    "social":['socio','costume']
}
```

```python
def genres_preprocessing(genres_dict, genre):
    genre_d = dict()
    
    # slash
    splitted = re.split("[,/]", genre)
    if len(splitted) != 1:
        genre_d['genre1'] = splitted[0].strip()
        genre_d['genre2'] = splitted[1].strip()
#         genre = splitted[0].strip(), splitted[1].strip()
    else:
        genre_d['genre1'] = genre 
    
    # trattini 
    for k,v in genre_d.items():
        splitted = re.split("[-—–]", genre)
        if (len(splitted) != 1) and (any(item in genres_dict.keys() for item in splitted)):
            genre_d['genre1'] = splitted[0].strip()
                #genre_d['genre2'] = splitted[1].strip()
    
    # spazi 
    for k,v in genre_d.items():
        splitted = re.split(" ", v.rstrip())
        if (len(splitted) != 1) and (any(item in genres_dict.keys() for item in splitted)):
            genre_d[k] = splitted[-1]
        elif (len(splitted) != 1) and (all(item in genres_dict.keys() for item in splitted)):
            genre_d[k] = splitted[0]
            genre_d['genre2'] = splitted[1]
    
    # conversione
    for key in genres_dict.keys():
        for value in genre_d.values():
            if value in genres_dict[key]:
                value = key
                
    return genre_d
```

```python
var = "sci-fi horror"
print(genres_preprocessing(conversion_dict, var))
```

```python
df['Genre2'] = df['Genre']
df['Genre2'] = df['Genre2'].apply(lambda x: genres_preprocessing(conversion_dict, x))
```

### IMDB

```python
df_imdb = pd.read_csv("../Imdb/imdb_dataset.csv",index_col=0)
df_imdb.columns = ['titleId','Title','startYear','genres']
```

```python
new_df = df[df.Title.isin(df_imdb.Title)]
new_df = new_df.dropna(subset=["Plot","Genre"]).merge(df_imdb, on='Title')
new_df = new_df.drop_duplicates(subset=['Title'])
```

```python
new_df
```

```python
mask = new_df.genres.apply(lambda x: any(g in x for g in ['Game-Show','Talk-Show','Mystery','\\N','News','Reality-TV']))
new_df = new_df[~mask]
new_df.shape
```

```python
new_df = new_df[['Title','Plot','genres']]
```

```python
import re

def genres_prep(genre):
    if 'Short' in genre:
        genre_final = ""
        for g in genre.split(","):
            if g.rstrip() != "Short":
                genre_final += g + ","
                
        genre_final = genre_final.rstrip(",")
    
    else:
        genre_final = genre
    
    return genre_final
```

```python
genres_prep("wdadwa,Short,Western")
```

```python
new_df = new_df[new_df.genres != 'Short']
new_df['genres'] = new_df.genres.apply(genres_prep)
```
